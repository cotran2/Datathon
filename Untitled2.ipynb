```python
import pandas as pd
```

```python
Traffic =  pd.read_csv(r'D:\Documents\ML\Uber\DOT_Traffic_Speeds_NBE.csv')
```

```python
df_2 = pd.read_csv(r'D:\Documents\ML\Uber\DatathonWeather.csv')
```

```python
df_2.dropna()
df_2.sample(10)
```

```python
df_2['Date/Time'] = pd.to_datetime(df_2['date'], format="%m/%d/%Y")
df_2['DayOfWeekNum'] = df_2['Date/Time'].dt.dayofweek
df_2['DayOfWeek'] = df_2['Date/Time'].dt.weekday_name
df_2['MonthDayNum'] = df_2['Date/Time'].dt.day
df_2['Month'] = df_2['Date/Time'].dt.month
df_2['Year'] = df_2['Date/Time'].dt.year
```

```python
df_2[['Month','precipitation']].groupby(['Month'],as_index = False).sum()
```

```python
import matplotlib.pyplot as plt
df_weekdays = df_2.pivot_table(index=['Month'],
                                    values = 'precipitation',
                                  aggfunc='sum')
df_weekdays.plot(kind='bar', figsize=(8,6),stacked=True)
plt.ylabel('Precipitation(inches)')
plt.title('Amount of precipitation(inches) by month');
plt.show()
```

```python
pd.crosstab(df_2.Month,df_2.MonthDayNum,margins=True).style.background_gradient(cmap='summer_r')
```
